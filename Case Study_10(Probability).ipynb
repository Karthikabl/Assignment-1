```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in data.columns:
    data[column] = le.fit_transform(data[column])
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(data.corr(), annot=True)
plt.tight_layout()
plt.show()
```

```python
# Split and Train the data
x= data.drop(['class'], axis=1)
y = data['class']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.20)
```

```python
# Classification Methods
```

```python
# 1. Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred_lr= lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
print('Recall          : ', recall_score(y_test,y_pred_lr))
print('f1 Score        : ', f1_score(y_test,y_pred_lr))
print('Precision       : ', precision_score(y_test,y_pred_lr))
print('Accuracy        : ', accuracy_score(y_test,y_pred_lr))
print("Confuion Matrix : \n ", confusion_matrix(y_test, y_pred_lr))
print("Test Accuracy   : {}%".format(round(lr.score(x_test, y_test)*100,2)))
```

```python
# 2. KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
acc_values = []
neighbors = np.arange(1,10)
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=k, metric = 'minkowski')
    classifier.fit(x_train, y_train)
    y_pred_knn = classifier.predict(x_test)
    acc = accuracy_score(y_test, y_pred_knn)
    acc_values.append(acc)
print(acc_values)
```

```python
plt.plot(neighbors,acc_values,'o-', color='teal')
plt.xlabel('K Value')
plt.ylabel('accuracy')
```

```python
classifier = KNeighborsClassifier(n_neighbors=1,metric='minkowski')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
print("Accuracy(K Value=1) : ",accuracy_score(y_test,y_pred))
```

```python
print('Recall          : ', recall_score(y_test,y_pred_knn))
print('f1 Score        : ', f1_score(y_test,y_pred_knn))
print('Precision       : ', precision_score(y_test,y_pred_knn))
print('Accuracy        : ', accuracy_score(y_test,y_pred_knn))
print("Confuion Matrix : \n", confusion_matrix(y_test, y_pred_knn))
print("Test Accuracy   : {}%".format(round(classifier.score(x_test, y_pred_knn)*100,2)))
```

```python
# 3. Decision Tree

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
y_pred_dt = dtc.predict(x_test)
```
